```python
#To load libraries
import requests 
from bs4 import BeautifulSoup 
from collections import Counter 
import pandas as pd
```

```python
phone_link = 'https://ephi.gov.et/about-us/regions/'
reply = requests.get(phone_link) #saving the result of the request in a variable called reply.
```

```python
print(reply.status_code) #I can use status_code to check if my request is successful
#The code print(variable name.status_code) is typically used in Python, often with libraries like requests to interact with web servers. The reply.status_code returns the HTTP status code of the response.
#When reply.status_code equals 200, it means that the request was successful, and the server returned a valid response. Essentially, 200 is the standard HTTP status code indicating that everything went as expected.
```

```python
'''After we checked that the request was successful, we use an attribute called
.text on the response data type to extract the html code from the page we
requested.'''
reply_str = reply.text
type(reply_str)
```

```python
print(reply_str)
```

```python
soup = BeautifulSoup(reply_str)
type(soup)
```

```python
regions = soup.find_all('tr')
print(regions)
```

```python
regions[10]
```

```python
single_reg = regions[10].find_all('td')
print(single_reg)
```

```python
for region in single_reg:
    print(region.text)
```

```python
result = []
for region in single_reg:
    print(region.text)
    result.append(region.text)
    print(result)
```

```python
total_results = []
for region in regions:
    result = []
    region_parsed = region.find_all('td')

    for region in region_parsed:
        result.append(region.text)
    total_results.append(result)
```

```python
print(total_results)
```

```python
#total_results = total_results[1:-1]
```

```python
df = pd.DataFrame(data = total_results, columns = ['sn', 'regional_bureau', 'phone'])
```

```python
df.head()
df
```

```python
#To remove unnecessary string from the phone number 
df['phone'] = df['phone'].str.replace("/td>", "") # Print the cleaned phone 
print(df['phone'])
```

```python
# Function to generate phone1, phone2, and phone3 from phone
def generate_phones(phone):
    parts = phone.split('/')
    phone1 = parts[0]
    phone2 = phone3 = ""
    if len(parts) == 2:
        phone2 = phone1.rsplit('-', 1)[0] + '-' + parts[1]
    elif len(parts) == 3:
        phone2 = phone1.rsplit('-', 1)[0] + '-' + parts[1]
        phone3 = phone1.rsplit('-', 1)[0] + '-' + parts[2]
    return phone1, phone2, phone3

# Apply the function to generate phone1, phone2, and phone3
df[['phone1', 'phone2', 'phone3']] = df['phone'].apply(generate_phones).apply(pd.Series)

# Replace phone2 and phone3 with empty strings based on conditions
df.loc[df['phone2'] == df['phone1'], 'phone2'] = ''
df.loc[df['phone3'] == df['phone1'], 'phone3'] = ''
df.loc[df['phone2'] == df['phone'], 'phone2'] = ''
df.loc[df['phone3'] == df['phone'], 'phone3'] = ''
```

```python
df
```

```python
# Drop the original 'phone' column 
df = df.drop(columns=['phone'])
```

```python
#To save to CSV file
export_file_name = 'Regional_healthbureau_phone.csv'
df.to_csv(export_file_name, index = False)
```
